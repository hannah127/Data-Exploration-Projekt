# Evaluation

In this section the ML models that have been trained in the notebook "Training_ML_Models" will be evaluated with the help of a confusion matrix, f1 score, accurancy, precision, recall and a ROC curve.

For our Use-Case a high precision is more relevant than a high recall. The reason for this is that we do not want to predict students that would graduate as dropouts (minimize false positives - dropout ) to avoid that these students get less opportunities in their life.

However, in order to have a complete evaluation of the models we will include both metrics.

## Imports

```python
import importnb
import matplotlib.pyplot as plt
from sklearn import metrics
import numpy as np
from sklearn.metrics import accuracy_score, confusion_matrix, ConfusionMatrixDisplay, classification_report
```

```python
with importnb.Notebook():
    from Data_Cleanup import df_filtered
    from Training_ML_Models import X_train, X_test, y_test, y_pred_knn, y_pred_knn_tuned, y_pred_rf, y_pred_rf_tuned, y_pred_dt, y_pred_dt_tuned, y_pred_gb, model_knn, model_knn_tuned, model_rf, model_rf_tuned, model_dt, model_dt_tuned, model_gb

df_filtered.head()
```

## 1. kNN

### 1.1 Evaluation of Untuned Model

```python
accuracy_model_knn = accuracy_score(y_test, y_pred_knn)
conf_matrix_knn = confusion_matrix(y_test, y_pred_knn)
class_report_knn = classification_report(y_test, y_pred_knn)

print(f'Accuracy: {accuracy_model_knn}')
print('Confusion Matrix:')
print(conf_matrix_knn)
print('Classification Report:')
print(class_report_knn)
```

The precision for class 0 (Dropout) is 92%, but the recall is 75%. This indicates that the untuned knn model classifies most of the dropouts correctly but misses some (classifies them as graduates).

The precison for class 1 (Graduate) is 85%, but the recall is 96%. This indicates that it correctly identifies most of the students that graduate. 

```python
ConfusionMatrixDisplay(confusion_matrix=conf_matrix_knn).plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
y_pred_probabiblity_knn = model_knn.predict_proba(X_test)[:,1]
fpr, tpr, thresholds = metrics.roc_curve(y_test,  y_pred_probabiblity_knn)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(fpr, tpr, label='Decision Tree Classification', color = 'firebrick')
ax.set_title('ROC Curve')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
plt.box(False)
ax.legend()
```

### 1.2 Evaluation of Tuned Model

```python
accuracy_model_knn_tuned = accuracy_score(y_test, y_pred_knn_tuned)
conf_matrix_knn_tuned = confusion_matrix(y_test, y_pred_knn_tuned)
class_report_knn_tuned = classification_report(y_test, y_pred_knn_tuned)

print(f'Accuracy: {accuracy_model_knn_tuned}')
print('Confusion Matrix:')
print(conf_matrix_knn_tuned)
print('Classification Report:')
print(class_report_knn_tuned)
```

```python
ConfusionMatrixDisplay(confusion_matrix=conf_matrix_knn_tuned).plot(cmap=plt.cm.Blues)
plt.title("Confusion Matrix")
plt.show()
```

```python
y_pred_probabiblity_knn_tuned = model_knn_tuned.predict_proba(X_test)[:,1]
fpr_tuned, tpr_tuned, thresholds_tuned = metrics.roc_curve(y_test,  y_pred_probabiblity_knn_tuned)

fig, ax = plt.subplots(figsize=(6,6))
ax.plot(fpr_tuned, tpr_tuned, label='Decision Tree Classification', color = 'firebrick')
ax.set_title('ROC Curve')
ax.set_xlabel('False Positive Rate')
ax.set_ylabel('True Positive Rate')
plt.box(False)
ax.legend()
```

The tuned kNN model performed better with an accurancy of 0.886 compared to the untuned model with an accurancy of 0.875. Also the precision of the tuned model for both classes is higher.

## 5. Result

To summarize the results from above which ML-models performed best:
- kNN: tuned model with an accurancy of 88,6%
- random forest: untuned model with an accurancy of 90,4%
- decision tree: tuned model with an accurancy of 88,7%
- naive bayes: untuned model with 86,3%

All models performed good on the dataset because they are all fitted to solve classification tasks. However, some models still performed a little bit worse than others.

Naive bayes performed worst out of all models. Why?
- we used guassian naive bayes
- assumes that each feature is normally distributed
- not the case: only the distribution of the feature admission grade can be counted as normally distributed
- dataset also has a lot of categorical features 

Decision Tree and kNN led to similar results. Why?
- good results however decision tree performs worse than random forest because there is a higher chance in overfitting because the tree can grow very much in complexity. This complex tree can perform worse on the testing data
- dnecision tree: easiert interpretable through the graphics
- Why do both perform similar? Both can handle categorical (if they are encoded) and numerical variables
- medium sized dataset -> performance would differ if the dataset would be larger
- all features used for the classification are relevant to the target variable -> both models can effectively capture the patterns and relationships within the data, leading to a similar performance of the model

The untuned Random Forest model performed best on the dataset. Why?
- is an ensemble of decision trees
- ensembles allow error reduction => results in higher accurancy
- combines bagging (sample diverse datasets from undelying data) and random attribute selection
- Is less likey to overfit than a decision tree because it combines the prediction of multiple decision trees which reduces the variance of the model 
- lower variance results in a better generalization thus better performance on an unnown dataset
